```
# import libraries
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
import tqdm
```

# Import and process the data

```
# import dataset (comes with colab!)
data = np.loadtxt(open('sample_data/mnist_train_small.csv','rb'),delimiter=',')

# don't need labels!
data = data[:,1:]

# normalize the data to a range of [0 1]
dataNorm = data / np.max(data)

# convert to tensor
dataT = torch.tensor( dataNorm ).float()
```

# Create the DL model

```
# create a class for the model
def createTheMNISTAE(encUnits,botUnits):

  class aenet(nn.Module):
    def __init__(self):
      super().__init__()

      ### input layer
      self.input = nn.Linear(784,encUnits)
      
      ### encoder layer
      self.enc = nn.Linear(encUnits,botUnits)

      ### latent layer
      self.lat = nn.Linear(botUnits,encUnits)

      ### decoder layer
      self.dec = nn.Linear(encUnits,784)

    # forward pass
    def forward(self,x):
      x = F.relu( self.input(x) )
      x = F.relu( self.enc(x) )
      x = F.relu( self.lat(x) )
      y = torch.sigmoid( self.dec(x) )
      return y
  
  # create the model instance
  net = aenet()
  
  # loss function
  lossfun = nn.MSELoss()

  # optimizer
  optimizer = torch.optim.Adam(net.parameters(),lr=.001)

  return net,lossfun,optimizer
```

```
# test the model with a bit of data
net,lossfun,optimizer = createTheMNISTAE(250,50)

X = dataT[:5,:]
yHat = net(X)

print(X.shape)
print(yHat.shape)
```

```
# let's see what the model did!
fig,axs = plt.subplots(2,5,figsize=(10,3))

for i in range(5):
  axs[0,i].imshow(X[i,:].view(28,28).detach() ,cmap='gray')
  axs[1,i].imshow(yHat[i,:].view(28,28).detach() ,cmap='gray')
  axs[0,i].set_xticks([]), axs[0,i].set_yticks([])
  axs[1,i].set_xticks([]), axs[1,i].set_yticks([])
  
plt.suptitle('Yikes!!!')
plt.show()
```

# Create a function that trains the model

```
randomidx = np.random.choice(dataT.shape[0],size=32)
dataT.shape
np.arange(0,32)
```

```
def function2trainTheModel(encunits,bottunits):

  # number of epochs
  numepochs = 20
  
  # create a new model
  net,lossfun,optimizer = createTheMNISTAE(encunits,bottunits)

  # initialize losses
  losses = torch.zeros(numepochs)


  # loop over epochs
  for epochi in range(numepochs):
    batchLoss = []

    for idx in range(0, dataT.shape[0],32):

      # select a random set of images
      if idx+31 > dataT.shape[0]-1:
        break
      idx = np.arange(idx,idx+31)
      X = dataT[randomidx,:]

      # forward pass and loss
      yHat = net(X)
      loss = lossfun(yHat,X)

      # backprop
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
      batchLoss.append(loss.item())

    # losses in this epoch
    losses[epochi] = np.mean(batchLoss)
  # end epochs

  # function output
  return losses,net
```

# Run the model and show the results!

```
# train the model (90s)
losses,net = function2trainTheModel(250,50)
print(f'Final loss: {losses[-1]:.4f}')

# visualize the losses
plt.plot(losses,'.-')
plt.xlabel('Epochs')
plt.ylabel('Model loss')
plt.title('OK, but what did it actually learn??')
plt.show()
```

```
# Repeat the visualization when testing the model
X = dataT[5:10,:]
yHat = net(X)


# let's see what the model did!
fig,axs = plt.subplots(2,5,figsize=(10,3))

for i in range(5):
  axs[0,i].imshow(X[i,:].view(28,28).detach() ,cmap='gray')
  axs[1,i].imshow(yHat[i,:].view(28,28).detach() ,cmap='gray')
  axs[0,i].set_xticks([]), axs[0,i].set_yticks([])
  axs[1,i].set_xticks([]), axs[1,i].set_yticks([])
  
plt.suptitle('Disregard the yikes!!!')
plt.show()
```
